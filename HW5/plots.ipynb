```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
sgd_memm = []
with open("predictions/adam_memm.txt") as f:
    for n,line in enumerate(f):
        if n%2!=0:
            try:
                line = line.strip()
                line = line.split("is ")[1]
                loss = float((line))
                sgd_memm.append(loss)
            except:
                continue
adam_memm = []
with open("predictions/sgd_memm.txt") as f:
    for n,line in enumerate(f):
        if n%2!=0:
            try:
                line = line.strip()
                line = line.split("is ")[1]
                loss = float((line))
                adam_memm.append(loss)
            except:
                continue
```

```python
x1 = range(len(sgd_memm))
x2 = range(len(adam_memm))
```

```python
plt.plot(x1, sgd_memm,label="SGD", linewidth=2)
plt.plot(x2, adam_memm,label="Adam", linewidth=2)
plt.legend()
plt.xlabel("Iterations (100)")
plt.ylabel("Loss")
plt.show()
```
